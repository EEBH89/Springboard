**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Get list of columns names, not including borough and ID
month_col_names = properties.columns[2:304]
print(month_col_names)
```

```python
# use month_col_names to melt dataframe
properties_m = pd.melt(properties, id_vars = ['boroughs', 'ID'], value_vars=month_col_names, value_name='average price', var_name="month")
properties_m.tail(10)
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# check data types of columns
properties_m['average price']
properties_m['month']
```

```python
# Change average price column from object to floating point
properties_m['average price'] = properties_m['average price'].astype(float)
properties_m['average price']
```

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
# identify and list non-borough names, filter dataframe to drop those, check boroughs again
invalid_boroughs=['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS','EAST OF ENGLAND', 'LONDON', 'SOUTH EAST','SOUTH WEST', 'England']

properties_clean = properties_clean[~properties_clean['boroughs'].isin(invalid_boroughs)]
properties_clean['boroughs'].unique()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# subset for borough = Lambeth, plot
properties_clean[properties_clean['boroughs']=='Lambeth'].plot(x='month', y='average price', kind='line')
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
# create year column
properties_clean['year'] = properties_clean['month'].apply(lambda t:t.year)
properties_clean.head()

# group by borough and year, with mean of average price, reset index
df = properties_clean.groupby(by=['boroughs', 'year']).mean()
df= df.reset_index()
df.head()
```

```python
# explore general trends of each borough as groups
df_p = properties_clean.groupby(by=['boroughs', 'year']).mean()
df_p = df_p.reset_index()


b_groups = df_p.groupby('boroughs')

fig, ax = plt.subplots()
for name, group in b_groups:
    ax.plot(group['year'], group['average price'])

plt.show()
```

```python
# identify least expensive borough in 2020, 1995

df_sorted = df_p[df_p['year'] == 1995].sort_values('average price')
df_sorted.head(20)

# least expensive borough is Barking & Dagenham $300,000 in 2020
```

```python
#identify most expensive borough

df_sorted = df_p[df_p['year'] == 2020].sort_values('average price', ascending=False)
df_sorted.head(20)

# most expensive borough is Kensington & Chelsea 1.2 million in 2020
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear inmind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# create price ratio function for 1998 to 2018

def price_ratio(arg):
    avg_price_98 = float(arg['average price'][arg['year'] == 1998])
    avg_price_18 = float(arg['average price'][arg['year'] == 2018])
    ratio = avg_price_98 / avg_price_18
    return ratio

# test function

test = price_ratio(df[df['boroughs']=='Lambeth'])
print(test)
```

```python
# run price ratio function over df for new dataframe

ratios = {}

for item in df['boroughs'].unique():
    borough = df[df['boroughs']==item]
    ratio = price_ratio(borough)
    ratios[item] = ratio
    
print(ratios)
```

```python
# turn ratios dict into dataframe

df=pd.DataFrame([ratios])
df.head()
```

```python
# transpose 

df = df.transpose()
df.head()
```

```python
# reset index
df=df.reset_index()
df.head()
```

```python
# rename columns
df.rename(columns={'index': 'borough', 0:'price ratio'}, inplace=True) 
df.head()
```

```python
df.sort_values(by='price ratio', ascending = False)
```

```python
#top 10 increasing boroughs
top_ten = df.sort_values(by='price ratio', ascending = False).loc[:10]
top_ten
```
